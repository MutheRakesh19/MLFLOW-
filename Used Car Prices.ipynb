```python
mean=df['Seats'].mean()
df['Seats']=df['Seats'].fillna(mean)
```

```python
df['Mileage'].fillna(df['Mileage'].mode()[0], inplace=True)
```

```python
df['Engine'].fillna(df['Engine'].mode()[0], inplace=True)
df['Power'].fillna(df['Power'].mode()[0], inplace=True)
df['Seats'].fillna(df['Seats'].mode())
df['Price'].fillna(df['Price'].mean())
```

```python
df.drop('New_Price',axis=1,inplace=True)
```

```python
df.info()
```

```python
numerical=df.select_dtypes(include=['int64','float64'])
categorical=df.select_dtypes(include='object')
```

```python
numerical
```

```python
for i in numerical:
    sns.boxplot(numerical[i])
    plt.show()
```

```python
# scaling the numerical variable
from sklearn.preprocessing import StandardScaler
# Now apply StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numerical)
```

```python
scaled_data=pd.DataFrame(scaled_data,columns=numerical.columns)
scaled_data.head()
```

```python
categorical
```

```python
categorical.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
categorical_encode=categorical.apply(label.fit_transform)
categorical_encode.head()
```

```python
df1=pd.concat([scaled_data,categorical_encode],axis=1)
```

```python
df1.head()
```

```python
x=df1.drop('Price',axis=1)
y=df1['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LinearRegression 
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import  r2_score
r2_score=r2_score(y_test,y_pred)
```

```python
r2_score
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_predict = lr.predict(x_test)
```

```python
 from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
y_predict = (y_predict > 0.5).astype(int)  # Assuming binary classification
# accuracy_score(y_test, y_predict)
# confusion_matrix(y_test, y_predict)
print("Unique values in y_test:", np.unique(y_test))
print("Unique values in y_predict:", np.unique(y_predict))
```
