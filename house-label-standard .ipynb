### Data preprocesing

#### Check the Outliers with boxplot

```python
# identify the ouliers with box plot
for i in numerical_cols:
    sns.boxplot(data=df,x=i)
    plt.show()
```

#### Handiling the Outliers with Wisker

```python
# remove the outliars with wisler
def wisker(col):
    q1=numerical_cols[col].quantile(0.25)
    q3=numerical_cols[col].quantile(0.75)
    iqr=q3-q1
    lower=q1-(1.5*iqr)
    upper=q3+(1.5*iqr)
    return lower,upper
    
```

```python
for i in numerical_cols:
  lower,upper=wisker(i)
  numerical_cols[i]=np.where(numerical_cols[i]<lower,lower,df[i])
  numerical_cols[i]=np.where(numerical_cols[i]>upper,upper,df[i])
```

### Import the MLFLOW

```python
import mlflow
```

```python
import os
os.environ['GIT_PYTHON_REFRESH']='quiet'
os.environ['MLFLOW_TRACKING_URI']='file:///C:/Raki'
```

```python
mlflow.set_experiment('house-label-std')
mlflow.set_experiment_tag('house-pred','ex2')
```

```python
with mlflow.start_run():
    # Model object initialization
    from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
    from sklearn.neighbors import KNeighborsRegressor
    from sklearn.svm import SVR
    linear = LinearRegression()
    RandFor = RandomForestRegressor(n_estimators=3000, max_depth=6)
    GBM = GradientBoostingRegressor(n_estimators=2000, max_depth=6, learning_rate=0.01, min_samples_split=5)
    # xgb_model = xgb.XGBRegressor(objective='reg:linear', n_estimators=1000, max_depth=6, learning_rate=0.1,
                                 # colsample_bytree=0.3, alpha=20)
    knn = KNeighborsRegressor(n_neighbors=100)
    svr = SVR(kernel='rbf', C=1.0, epsilon=0.1, gamma='scale')

    rf_n_estimators = 3000

    # Logging parameters
    mlflow.log_param('n_estimators', rf_n_estimators)
    mlflow.log_param('rf_max_depth', 6)
    mlflow.log_param('gbm_n_estimators', 2000)
    mlflow.log_param('gbm_max_depth', 6)
    mlflow.log_param('gbm_learning_rate', 0.01)
    # mlflow.log_param('xgb_objective', 'reg:linear')
    # mlflow.log_param('xgb_n_estimators', 1000)
    # mlflow.log_param('xgb_max_depth', 6)
    # mlflow.log_param('xgb_learning_rate', 0.1)
    # mlflow.log_param('xgb_colsample_bytree', 0.3)
    # mlflow.log_param('xgb_alpha', 20)
    mlflow.log_param('knn_n_neighbors', 100)
    mlflow.log_param('svr_kernel', 'rbf')
    mlflow.log_param('svr_C', 1.0)
    mlflow.log_param('svr_epsilon', 0.1)
    mlflow.log_param('svr_gamma', 'scale')

    # Training models
    linear.fit(x_train, y_train)
    rf = RandFor.fit(x_train, y_train)
    gbm = GBM.fit(x_train, y_train)
    # xgb_model.fit(x_train, y_train)
    knn.fit(x_train, y_train)
    svr.fit(x_train, y_train)

    # Training scores
    training_score_linear = linear.score(x_train, y_train)
    training_score_rf = rf.score(x_train, y_train)
    training_score_gbm = gbm.score(x_train, y_train)
    # training_score_xgb = xgb_model.score(x_train, y_train)
    training_score_knn = knn.score(x_train, y_train)
    training_score_svr = svr.score(x_train, y_train)

    # Logging training scores
    mlflow.log_metric('linear_training_score', training_score_linear)
    mlflow.log_metric('rf_training_score', training_score_rf)
    mlflow.log_metric('gbm_training_score', training_score_gbm)
    # mlflow.log_metric('xgb_training_score', training_score_xgb)
    mlflow.log_metric('knn_training_score', training_score_knn)
    mlflow.log_metric('svr_training_score', training_score_svr)

    # Value prediction
    y_pred_linear = linear.predict(x_test)
    y_pred_rf = rf.predict(x_test)
    y_pred_gbm = gbm.predict(x_test)
    # y_pred_xgb = xgb_model.predict(x_test)
    y_pred_knn = knn.predict(x_test)
    y_pred_svr = svr.predict(x_test)

    # Prediction evaluations
    testing_score_linear = r2_score(y_test, y_pred_linear)
    testing_score_rf = r2_score(y_test, y_pred_rf)
    testing_score_gbm = r2_score(y_test, y_pred_gbm)
    # testing_score_xgb = r2_score(y_test, y_pred_xgb)
    testing_score_knn = r2_score(y_test, y_pred_knn)
    testing_score_svr = r2_score(y_test, y_pred_svr)

    # Logging evaluation scores
    mlflow.log_metric('linear_testing_score', testing_score_linear)
    mlflow.log_metric('rf_testing_score', testing_score_rf)
    mlflow.log_metric('gbm_testing_score', testing_score_gbm)
    # mlflow.log_metric('xgb_testing_score', testing_score_xgb)
    mlflow.log_metric('knn_testing_score', testing_score_knn)
    mlflow.log_metric('svr_testing_score', testing_score_svr)

    # Logging models
    mlflow.sklearn.log_model(linear, 'linear-regression-model')
    mlflow.sklearn.log_model(rf, 'random-forest-model')
    mlflow.sklearn.log_model(gbm, 'gradient-boosting-model')
    # mlflow.sklearn.log_model(xgb_model, 'xgboosting-model')
    mlflow.sklearn.log_model(knn, 'knn-model')
    mlflow.sklearn.log_model(svr, 'svr-model')
```
